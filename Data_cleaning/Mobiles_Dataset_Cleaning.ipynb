```python
import pandas as pd
dataset = pd.read_csv('Mobiles_Dataset.csv', encoding='ISO-8859-1')
print(dataset.head())
```

```python
dataset = dataset.drop(['Launched Price (Pakistan)', 'Launched Price (India)','Launched Price (China)','Launched Price (Dubai)'], axis=1)
```

```python
print(dataset.head())
```

```python

dataset.columns = [col.strip().replace('\\', '') for col in dataset.columns]

# Verify the columns are correct
expected_columns = [
    'Company Name', 'Model Name', 'Mobile Weight', 'RAM', 'Front Camera',
    'Back Camera', 'Processor', 'Battery Capacity', 'Screen Size',
    'Price (USA)', 'Launched Year'
]
if list(dataset.columns) != expected_columns:
    print("Warning: Columns may not match expected:", dataset.columns)

# Verify shape (should be 930, 11)
print("Shape:", dataset.shape)
```

```python
# Add Cambodia price (KHR) with VAT
exchange_rate = 4000  #assume that 1$ is 4000 riel
vat_rate = 0.155  # 10% VAT
dataset['Price (Cambodia, KHR)'] = (dataset['Launched Price (USA)'] * exchange_rate) * (1 + vat_rate)

# Round prices for readability
dataset['Price (Cambodia, KHR)'] = dataset['Price (Cambodia, KHR)'].round().astype(int)
```

```python
dataset
```

```python
dataset = dataset[dataset['Company Name'].str.lower() != 'motorola']
```

```python
dataset.isnull().sum()
```

```python
dataset = dataset[dataset['Company Name'].str.lower() != 'poco']
```

```python
dataset.reset_index(drop=True, inplace=True)
```
